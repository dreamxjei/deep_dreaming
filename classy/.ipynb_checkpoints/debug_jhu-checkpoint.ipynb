```python
import os
import numpy as np

import matplotlib.pyplot as plt
from PIL import Image
from skimage import io, color

import torch
from torchvision import transforms
```

```python
input_dir = r'risser_data'
```

```python
imgtransform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224, 224)),
    transforms.RandomRotation((-5, +5)),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.25),
    transforms.ToTensor()
])
```

### tha

```python
# control: look at THA data

idir = '~/work2/mamba/scoliosis/risser/archive/dataset_tha/train/no_THA'
ifile = os.path.join(idir, 'C7FF411.png')

raw = io.imread(ifile)
rawshape = np.shape(raw); print('raw shape:',rawshape)
if len(rawshape) == 3 and rawshape[2] > 3:  # for cases (x,y,4)
    raw = raw[:,:,3]  # was [:,:,3] but it's just blank!
    raw = color.gray2rgb(raw)
    print('raw shape after conditional:',raw.shape)
    
# raw = color.gray2rgb(raw)  # somehow this is OK

# square crop from bottom
ydim = rawshape[1]
xmax = rawshape[0]
xmin = xmax - ydim
# img = raw[xmin:xmax, :]; print('shape after crop:',np.shape(img))

img = raw
img_transformed = imgtransform(img)
print(np.shape(img_transformed))
# print('uniques after transform:',np.unique(img_transformed))

plt.figure()
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(img_transformed[0,:,:])
```

# skimage for all in train, risser4

```python
# check out risser=5

idir = os.path.join(input_dir,'4')
filenames = os.listdir(idir)
filenames.sort()

for filename in filenames:
    print('processing',filename)
    ifile = os.path.join(idir, filename)

    raw = io.imread(ifile)
    rawshape = np.shape(raw); print('raw shape:',rawshape)
    if len(rawshape) == 3 and rawshape[2] > 3:  # for cases (x,y,4)
        raw = raw[:,:,0]  # was [:,:,3] but it's just blank!
        raw = color.gray2rgb(raw)
        print('raw shape after conditional:',raw.shape)

    # raw = color.gray2rgb(raw)  # somehow this is OK

    # square crop from bottom
    ydim = rawshape[1]
    xmax = rawshape[0]
    xmin = xmax - ydim
    img = raw[xmin:xmax, :]; print('shape after crop:',np.shape(img))

    img_transformed = imgtransform(img)
    print(np.shape(img_transformed))
    # print('uniques after transform:',np.unique(img_transformed))

    plt.figure()
    plt.subplot(1,2,1)
    plt.imshow(img)
    plt.subplot(1,2,2)
    plt.imshow(img_transformed[0,:,:])
    plt.show()
```

### save cropped images for CAM generation

```python
# check out risser=5

idir = os.path.join(input_dir,'5')
filenames = os.listdir(idir)
filenames.sort()

for filename in filenames:
    print('processing',filename)
    ifile = os.path.join(idir, filename)

    raw = io.imread(ifile)
    rawshape = np.shape(raw); print('raw shape:',rawshape)
    if len(rawshape) == 3 and rawshape[2] > 3:  # for cases (x,y,4)
        raw = raw[:,:,0]  # was [:,:,3] but it's just blank!
        raw = color.gray2rgb(raw)
        print('raw shape after conditional:',raw.shape)

    # raw = color.gray2rgb(raw)  # somehow this is OK

    # square crop from bottom
    ydim = rawshape[1]
    xmax = rawshape[0]
    xmin = xmax - ydim
    img = raw[xmin:xmax, :]; print('shape after crop:',np.shape(img))
    
    # save image for CAM
    

    img_transformed = imgtransform(img)
    print(np.shape(img_transformed))
    # print('uniques after transform:',np.unique(img_transformed))

    plt.figure()
    plt.subplot(1,2,1)
    plt.imshow(img)
    plt.subplot(1,2,2)
    plt.imshow(img_transformed[0,:,:])
    plt.show()
```
